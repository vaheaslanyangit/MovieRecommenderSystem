# Text Cleaning (NLP Text Cleaning steps with NLTK)


```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

def clean_text(text):
    # Check if text is not a string
    if not isinstance(text, str):
        return ""  # or some other default value like 'NaN' or None

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation but keep digits
    text = re.sub(r'[^\w\s\d]', '', text)

    # Tokenize
    words = word_tokenize(text)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    # Lemmatize
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]

    # Join the words back together
    text = ' '.join(words)
    return text
```

```python
new_data['tags_clean'] = new_data['tags'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# This line of code is initializing a CountVectorizer object 'cv' with a maximum of 10,000 features and English stop words.
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
# This line of code is fitting the CountVectorizer 'cv' to the 'tags_clean' column of the 'new_data' DataFrame and transforming the text data into a numerical vector representation. The result is converted to a NumPy array and assigned to the variable 'vector'.
vector = cv.fit_transform(new_data['tags_clean'].values.astype('U')).toarray()
```

```python
new_data['tags']
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data.info()
```

```python
# This line of code is calculating the similarity scores for the third movie (index 2) with all other movies, sorting them in descending order, and storing the sorted list of tuples in the 'distance' variable. Each tuple contains the index of the movie and its similarity score with the third movie.
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])

# This loop is iterating over the first five tuples in the 'distance' list (the five movies most similar to the third movie) and printing the title of each movie.
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
# This function recommends the top 5 similar movies. It first finds the index of the input movie in the 'new_data' DataFrame. Then, it calculates the similarity scores of this movie with all other movies, sorts them in descending order, and prints the titles of the top 5 most similar movies.
def recommend(movies):
    index = new_data[new_data['title'] == movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
